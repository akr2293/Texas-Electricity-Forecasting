```python
path = '../data/final_dfs'
files = os.listdir(path)
files
```

```python
df_2009 = pd.read_csv('../data/final_dfs/2009_clean.csv', index_col='date')
df_2010 = pd.read_csv('../data/final_dfs/2010_clean.csv', index_col='date')
df_2011 = pd.read_csv('../data/final_dfs/2011_clean.csv', index_col='date')
df_2012 = pd.read_csv('../data/final_dfs/2012_clean.csv', index_col='date')
df_2013 = pd.read_csv('../data/final_dfs/2013_clean.csv', index_col='date')
df_2014 = pd.read_csv('../data/final_dfs/2014_clean.csv', index_col='date')
df_2015 = pd.read_csv('../data/final_dfs/2015_clean.csv', index_col='date')
df_2016 = pd.read_csv('../data/final_dfs/2016_clean.csv', index_col='date')
df_2017 = pd.read_csv('../data/final_dfs/2017_clean.csv', index_col='date')
df_2018 = pd.read_csv('../data/final_dfs/2018_clean.csv', index_col='date')
df_2019 = pd.read_csv('../data/final_dfs/2019_clean.csv', index_col='date')
df_2020 = pd.read_csv('../data/final_dfs/2020_clean.csv', index_col='date')
```

```python
big_df = pd.concat([df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020], axis=0)
big_df
```

```python
big_df.plot(figsize=(20,12)).legend(title='Source', bbox_to_anchor=(1, 1));
```

```python
plt.figure(figsize=(20,12))

big_df.rolling(30)['Biomass'].mean().plot(kind='line')
big_df.rolling(30)['Coal'].mean().plot(kind='line')
big_df.rolling(30)['NGas'].mean().plot(kind='line')
big_df.rolling(30)['Hydro'].mean().plot(kind='line')
big_df.rolling(30)['Nuclear'].mean().plot(kind='line')
big_df.rolling(30)['Other'].mean().plot(kind='line')
big_df.rolling(30)['Solar'].mean().plot(kind='line')
big_df.rolling(30)['Wind'].mean().plot(kind='line')

plt.legend(['Biomass', 'Coal', 'NGas', 'Hydro', 'Nuclear', 'Other', 'Solar', 'Wind'])
```

```python
big_df.to_csv('../data/final_dfs/combined_final.csv')
```
